```python
import pyspark
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("test") \
    .getOrCreate()
```

```python
df = spark.read \
    .option("header", "true") \
    .csv("taxi_zone_lookup.csv")
```

```python
df.show()
```

```python
df.write.parquet('zones')
```

```python
spark.stop()
```
